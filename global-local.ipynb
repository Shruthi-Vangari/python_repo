```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    print(a/b)
def calculator():
    print("enter 1 for addition")
    print("enter 2 for subtraction")
    print("enter 3 for multiplication")
    print("enter 4 for division")
    op=eval(input("enter from 1-4:"))
    a=eval(input("enter a:"))
    b=eval(input("enter b:"))
    if op==1:
        add(a,b)
    elif op==2:
        sub(a,b)
    elif op==3:
        mul(a,b)
    elif op==4:
        div(a,b)
    else:
        print("enter valid num")
calculator()
```
